# Objectives

In Lesson 24 (conservation of energy), one of the few potential energies we could define was the gravitational potential energy $U_g$. At that point, though, we were working close to the surface of the Earth, so the gravitational field was taken to be a constant. Hopefully, this case built some intuition about what it means to have the potential energy of a conservative force. Raising a ball higher into the air, for example, increases the energy stored in the ball-Earth system. When we let go of the ball, this energy is converted into the kinetic energy, as the ball speeds up in its fall. The energy of the ball-Earth system will decrease as the ball gets closer to the center of the Earth.

Now we have talked about the more general form of the gravitational force in Lesson 25. This force is still conservative, but the force changes as the distance between the two objects changes. As we raise the ball in the air, there is an ever so slight decrease in the gravitational field at its position. Fortunately for us, this effect is not noticeable. However, if we were to launch the ball at *very* high speeds, so it traveled beyond the atmosphere, then the change in the field would give a different trajectory than with the constant field case. Thus, we need to modify our concept of the gravitational potential energy to take into account the fact that the gravitational force is not constant in magnitude.

However, much of the intuition we gained in the constant field case holds true. We will see, for example, that the gravitational potential energy will increase as two objects increase the distance. Thus, a rocket launched from the Earth's surface will gain gravitational potential energy, with a corresponding loss of kinetic energy. On the other hand, there may be some aspects of the new form of $U_g$ that take some getting used to! To motivate this, we will start with the form of universal gravitation, and show how it gives a distance-dependent equation for $U_g$. Then, we discuss the general properties of this potential energy, and show how we can calculate the speeds of rockets leaving the surface of planets.

Here are the objectives for this lesson:

* Define gravitational potential energy.
* Identify the reference radius for zero gravitational potential energy.
* Calculate the escape speed necessary to travel very far away from an object.

# Gravitational potential energy

Just as a reminder, we have from Lesson 25 that Newton's law of universal gravitation is given by

$$
    {\vec F}_{g, ij} = -\biggl( \frac{G m_i m_j}{r^2 _{ij}} \biggr) {\hat r}_{ij}
$$

where $r_{ij}$ is the distance between the two masses interacting gravitationally. Before we get into the matching form of the gravitational potential energy $U_g$, let's see why we can assume this is constant near the surface of the Earth. In particular, we can graph the gravitational field magnitude $g$ relatively near the Earth's surface, but maybe a little higher than we experience in everyday life.

Below is a graph of the gravitational acceleration $g$ due to the Earth versus the height $h$ above the Earth's surface. Near the surface, the acceleration is the well-known 9.8 m/s$^2$, but this decreases to about 9.0 m/s$^2$ at a height of about 250 km. This is a difference of about 8% at this distance.

![Graph of gravitational potential energy vs. height above the Earth's surface](../img/INT_AY20_MP2_L28_Fig01-GPE_vs_h_graph_small.png)

Notice a few things about this graph. The typical definition of the height of the Earth's atmosphere is about 100 km or so. So, even at that altitude, the magnitude of the gravitational field is rather close to that on the surface of the Earth. If we climbed the highest mountains, we would barely even notice the change! This is why we can assume the gravitational field is constant -- for almost all applications, it basically stays the same.

However, there eventually is a difference in $g$. Over the span of the graph, the curve looks much like a straight line, but if we kept going further, we would notice that it curves. This is because of the inverse square law nature of the gravitational force ($1/r^2$ in the denominator). Thus, we expect to see something similar with the gravitational potential energy $U_g$.

So how can we find $U_g$? First, remember that the change in the potential energy $U_F$ for a conservative force ${\vec F}$ is equal to the work $W_F$ done by that force over a displacement $\Delta {\vec r}$. Mathematically, this is

$$
    \Delta U_F = -W_F = -{\vec F} \cdot \Delta {\vec r}
$$

When we assumed the gravitational field $g$ was constant, then the change in gravitational potential energy would be $\Delta U_g = mg \Delta r_y$. Notice what is happening there: the force ($mg$) is multiplied by the vertical displacement $\Delta r_y$. Notice also that there is no special choice for which $r_y$ the potential energy should be zero. In other words, we were free to choose our "reference height" as the zero value for potential energy.

Now let's see how things are affected by a changing gravitational field. We can use the relationship between the potential energy and the work done by the force to get the form of $U_g$. Strictly speaking, we are about to do some bad math, but it turns out that it actually gives you the right answer! Specifically, let's naively say the new gravitational potential energy is given by (taking the scalar product)

$$
    \Delta U_g = -\biggl( \frac{Gm_i m_j}{r_{ij} ^2} \biggr) \Delta r_{ij}
$$

where $r_{ij}$ is the distance between the two masses $m_i$ and $m_j$. If we ignore the fact that it really says $\Delta r_{ij}$, not just $r_{ij}$, and let $\Delta r_{ij}$ cancel out one of the $r_{ij}$'s in the denominator, we get

$$
    U_g = - \frac{G m_i m_j}{r_{ij}}
$$

This is the right answer! Again, properly speaking we should go through the details of why the $\Delta$'s go away so nicely, but we will not worry too much about it here. Thus, we have the definition of **gravitational potential energy**. This is a scalar, with units of joules (J), just like all other forms of energy.

We should return to something said above, since it will be important later on. When the gravitational field was constant, we were free to pick our reference height, where $U_g = 0$. But we can't do that here -- it is already chosen for us! This situation is more like that of a spring, and elastic potential energy. In that case, there is a natural place for the reference, namely the equilibrium point $r = 0$ where the spring is neither stretched nor compressed. Obviously, if the spring is at its equilbrium point, there is no energy stored in the spring, so it makes sense that $U_{spr} = 0$ there.

What is the corresponding location for gravity? Note this is the same as asking how far do two objects have to be to store no gravitational energy due to their interaction. If they are close to each other, there is a relatively strong gravitational force between them, so this energy can be easily converted into kinetic energy, for example. However, if they get further apart, the gravitational force between them gets smaller, so there is less energy in their interaction. This energy goes to zero when they are "very far away" -- essentially at an infinite distance apart. This is the "reference" we will use for gravitational potential energy; the energy $U_g(r) = 0$ when $r \to \infty$. We will need this fact next to find the escape speed for a rocket leaving a planet.

# Escape speed

One of the first topics we covered in this class, back in Lesson 04, was what happens when you throw a ball up into the air -- how high does it go? Remember that this was using the assumption that the gravitational field was constant, as it is near the surface of the Earth. This field has very little variation over the heights we normally encounter during our daily routine.

But how would this change if something was moving very fast, and we had to worry about a decreasing gravitational field as we move away from the Earth? One way to do this would be to take into account the changing field. For example, you could write a Python program, where the position and velocity of the moving object is updated using an acceleration that depends on the height above the Earth's surface.

However, this is really overkill, since the changing field is "taken into account" by the gravitational potential energy. All we really need to do is use is conservation of mechanical energy, using the $U_g$ from this lesson, rather than the equation $mgh$ (which assumes a fixed gravitational field $g$). Then we can find how high an object moves if it has some starting speed at ground level.

To show how this works, let's go through the following example. The largest rocket ever built is the Saturn V, used to put US astronauts on the Moon. Suppose instead the $2.97 \times 10^6$ kg rocket is launched from the surface of the Earth into deep space. The Earth has a mass of $5.97 \times 10^{24}$ kg and a radius of 6,380 km; ignore the effect of the Sun and the other planets. We are going to find the necessary launch speed for the Saturn V to momentarily stop at the height of low Earth orbit (LEO) --  2000 km away above the surface of the Earth.

We are ignoring the effects of air resistance on the rocket, so $W_{NC} = 0$. Then we can use conservation of mechanical energy to solve for the final distance from the surface. At its launch, the rocket has both kinetic and gravitational potential energy; it starts at a distance $r = R_E = 6380$ km from the Earth's *center* with a speed $v_i$, so its initial mechanical energy is

$$
    E_i = \frac{1}{2} m_S v_i ^2 - \frac{GM_E m_S}{R_E}
$$

where $M_E$ is the mass of the Earth, and $m_S$ is the mass of the Saturn rocket. At the rocket's highest point, it *just* reaches LEO, so its final speed $v_f = 0$. This will be at a height $h = 2000$ km above the surface, so it is a distance $R_E + h$ away from the Earth's center. This gives the final mechanical energy as

$$
    E_f = -\frac{GM_E m_S}{R_E + h}
$$

Setting these equal to each other, notice that the mass $m_S$ is in every term, so it can be canceled out -- the mass of the rocket doesn't matter!

> **Ignoring the rocket mass**: You may be a little uncomfortable about the fact we can get rid of the Saturn rocket's mass. Does that mean we can make the rocket as big as we want, and it doesn't matter? Well, no, because I am really assuming that the mass $M_E$ of the Earth is much bigger than the mass $m_S$ of the rocket. If this were not true (imagine a binary star system, with two stars the same mass), then the masses we use would be more like an "averaged" mass, much like what we did in Lesson 09 when we talked about the center of mass. In this case, the masses we are using are essentially correct, and give a very small error, so I don't worry about it.

Dividing out by the mass of the rocket, this gives the energy equation

$$
    \frac{1}{2} v_i ^2 - \frac{GM_E}{R_E} = -\frac{GM_E}{R_E + h}
$$

The gravitational potential energy terms can be moved to the same side, then you can solve for $v_i$. This gives

$$
    v_i = \sqrt{2GM_E \biggl( \frac{1}{R_E} - \frac{1}{R_E + h} \biggr)}
$$

Plugging in the numbers from above, one finds the initial speed must be 5.46 km/s to just barely reach LEO.

Notice that we get a bonus by doing the algebra -- we can now find the necessary speed to reach any distance $h$ above the surface. In fact, we can figure out the speed needed to go "very far away", in other words, when $h \to \infty$. This gives the speed known as the **escape speed** $v_{esc}$: the speed required to travel an infinite distance away from another object. In practice, this just means a large distance, compared to the size of the object left (for example, the Earth). Starting with our last equation, and taking the limit of $h$ going to infinity, we have that

$$
    \lim_{h \to \infty} \frac{1}{R_E + h} = 0
$$

So the second term disappears, and the escape speed is

$$
    v_{esc} = \sqrt{\frac{2GM_E}{R_E}}
$$

We can now easily calculate the escape speed for the Earth as 11.2 km/s.

# Summary

We now have all of the tools needed to understand the motion of objects in the Solar System. In Lesson 25, we discussed Newton's law of universal gravitation; this is the interaction force between all bodies, large and small, moving around the Sun. Thinking about how this gravitational force provides a necessary centripetal force (Lesson 18) will allow us to start talking about the size of orbits, and the time needed to complete one.

In this lesson, we defined the gravitational potential energy, based on the gravitational force. Along with conservation of energy (Lesson 24), we can now relate the distance between the Earth and an orbiting satellite to its kinetic energy, and thus its speed. This will give us a relation between speed and distance. However, remember we also have conservation of angular momentum (Lesson 20), which relates distance and the *tangential* speed. Thus, between these two conservation principles, we can find the *velocity* of an orbiting object, given its distance away from a planet, because we can break the velocity into its radial and tangential components.

This overview shows us all the physics tools that we can use to understand orbital motion. With this perspective, we next head to Lesson 27, and show how these all come together in Kepler's laws of planetary motion.
