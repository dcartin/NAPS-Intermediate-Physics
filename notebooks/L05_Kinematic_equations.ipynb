# Kinematic equations

## Vector kinematic equations

For now, we examine the motion of an object assuming that the acceleration of the object is *constant*. This will allow us to derive a set of equations to use to solve for an unknown quantity. Later we will explore non-constant accelerations, using vPython programming and numerically calculating velocity and position. I emphasize the later point because, as a general rule, it is important to know when a particular equation "breaks". In other words, just like an engineer must find the safety margins for a particular structure or piece of equipment, you should know when an equation is valid or not. So, our key assumption here, to repeat, is that

$$
    {\vec a} = {\vec a}_{avg} = \textrm{constant}
$$

Later in this lesson, we will explore what happens when this assumption is *not* true.

We can build up all the kinematic equations from this one assumption, and the definitions of average velocity and acceleration. Starting from the definition of acceleration, we get our first equation; we will also use the symbol $t$ to represent $\Delta t$.

$$
	{\vec a} = \frac{\Delta {\vec v}}{\Delta t} = \frac{{\vec v}_f - {\vec v}_i}{t}
$$

This gives

$$
    {\vec v}_f = {\vec v}_i + {\vec a} t
$$

Note this *vector* equation is the same as an equation for *each* of the vector components:

$$
	\begin{eqnarray*}
		v_{f, x} &=& v_{i, x} + a_x t \\
		v_{f, y} &=& v_{i, y} + a_y t \\
		v_{f, z} &=& v_{i, z} + a_z t
	\end{eqnarray*}
$$

It is simpler to write the vector equation (three equations for the price of one!), but realize that it means the equation must be satisfied in each of the coordinate directions.

Another point is that we have been using this kinematic equation already when updating the velocity of an object in vPython. If we want to find the final velocity `ball.velocity` of an object `ball` at the end of a time step `dt`, with acceleration `ball.accel`, then 

```python
    ball.velocity = ball.velocity + ball.accel * dt
```

So hopefully you see the relationship between physics and computing when we construct simulations. However, we can go further in the simulations, even when the acceleration is *not* constant. In a case like that, the new velocity would be just an estimate, since the kinematic equation is no longer exactly accurate, and there would be some error in our computations. Reducing that error as much as possible is a key pursuit in computational physics.

Our next equation is going to give a relation between the displacement $\Delta {\vec r} = {\vec r}_f - {\vec r}_i$, the time interval $t$, and the initial and final velocities ${\vec v}_i$ and ${\vec v}_f$. Let's show this by reviewing some graph skills; we will specifically use the motion in the $x$ direction, but this argument works for the $y$ and $z$ directions as well. If we graph $v_x$ versus time for a constant acceleration, our graph looks like the one below.

![$x$ velocity vs. time graph for an object with a constant $a_x$](../img/INT_AY21_L05_Fig01-Vel-vs-t-graph_small.png)

I know that $a_x$ is constant, because the graph has a constant slope. To find the displacement, I need to find the area of the shaded region, defined by the time interval $t$. This can be divided into a rectangle and a triangle, so the area is 

$$
    \Delta r_x = v_{i, x} t + \frac{1}{2} (v_{f, x} - v_{i, x}) t = \frac{1}{2}(v_{i, x} + v_{f, x}) t
$$

Performing the same derivation for the other two directions, I can write this as a vector equation, namely,

$$
    \Delta {\vec r} = \frac{1}{2} ({\vec v}_i + {\vec v}_f) t
$$

## The scalar kinematic equations

We have one more set of equations to go, however. Although they may look like it, they cannot be written as a single vector equation, but there is one for each of the three coordinate axes $x, y$, and $z$. I will call these the **scalar kinematic equations**. Let's use the following problem to motivate these equations.

> **Problem:** Consider the running rabbit from a previous problem, with the motion shown in the figure below. Remember, the velocity of the rabbit at $t = 0$ is $(-9.10 \ \text{m/s}) {\hat x} + (7.20 \ \text{m/s}) {\hat y}$, and the rabbit's acceleration is $(43.6 \ \text{cm/s}^2) {\hat x} + (-62.0 \ \text{cm/s}^2) {\hat y}$. What is the *first* time $t > 0$ the rabbit has a position of $x = 0$?
>
> ![The motion of a rabbit](../img/Run-rabbit-run-graph_small.png)

There is more than one way to solve this problem, but one way to do it in one step is to use the $x$ component equation

$$
    \Delta r_x = v_{i, x} t + \frac{1}{2} a_x t^2
$$

However, there are two terms which have a time $t$ in them, so it is not a simple matter of solving for one of these terms, and then dividing; this method will not have a time only on one side of the equation, but both! Thus, you need something slightly more sophisticated -- here, you need to use the quadratic formula. Thus, you need to get the equation above into the standard form

$$
	0 = \frac{1}{2} a_x t^2 + v_{i, x} t - \Delta r_x
$$

and then solve for $t$. This gives (make sure you watch your signs!)

$$
		t = \frac{-v_{i, x} \pm \sqrt{v_{i, x} ^2 + 2 a_x \Delta r_x}}{a_x}
$$

Compare this to the equation we can get from the $x$ component of the equation ${\vec v}_f = {\vec v}_i + {\vec a} t$:

$$
	t = \frac{v_{f, x} - v_{i, x}}{a_x}
$$

These equations match up (and we don't have to worry about the $\pm$ sign) if

$$
    v_{f, x} ^2 = v_{i, x} ^2 + 2 a_x \Delta r_x
$$

Note these are the $x$ components of the vector, not the magnitudes; the same idea works for the $y$ and $z$ components as well. 

$$
    v_{f, y} ^2 = v_{i, y} ^2 + 2 a_y \Delta r_y \qquad
    v_{f, z} ^2 = v_{i, z} ^2 + 2 a_z \Delta r_z
$$

# Summary

This lesson expanded our mathematical tool kit to include the vector and scalar kinematic equations. For completeness, let's list them all here.

$$
\begin{eqnarray*}
    {\vec v}_f &=& {\vec v}_i + {\vec a} t \\
    \Delta {\vec r} &=& \frac{1}{2} ({\vec v}_i + {\vec v}_f) t \\
    \Delta {\vec r} &=& {\vec v}_i t + \frac{1}{2} {\vec a} t^2
\end{eqnarray*}
$$

and

$$
    v_{f, x} ^2 = v_{i, x} ^2 + 2 a_x \Delta r_x \qquad
    v_{f, y} ^2 = v_{i, y} ^2 + 2 a_y \Delta r_y \qquad
    v_{f, z} ^2 = v_{i, z} ^2 + 2 a_z \Delta r_z
$$

These are exact when the acceleration of an object is constant. Even when the acceleration is *not* constant, these kinematic equations are the basis for everything we will be doing in computational physics. You have already seen how the first two allow us to update the position and velocity of a moving object. Later in the course, in Lesson 20, you will see how the last three provide a helpful window into the idea of work and energy. Understanding how to use these in various applications will be very important for you!
